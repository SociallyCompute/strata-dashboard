## Ideal query - calculating yellow or green block
### With eventKey

```python
START_KEY = "dialogue:20:1"    
END_KEY   = "dialogue:19:46"

TARGET_KEYS = [
    "dialogue:18:99",  "dialogue:28:179", "dialogue:59:179",
    "dialogue:18:223", "dialogue:28:182", "dialogue:59:182",
    "dialogue:18:224", "dialogue:28:183", "dialogue:59:183"
]
```

```python
yellow = []
green = []

for pid in completed:

    start_doc = coll.find_one(
        {
            "game": "mhs",
            "playerId": pid,
            "eventType": "DialogueEvent",
            "eventKey": START_KEY
        },
        sort=[("timestamp", 1)],
        projection={"timestamp": 1}
    )

    start_ts = datetime.fromisoformat(start_doc["timestamp"].replace("Z", "+00:00"))

    end_doc = coll.find_one(
        {
            "game": "mhs",
            "playerId": pid,
            "eventType": "DialogueEvent",
            "eventKey": END_KEY,
            "timestamp": {"$gte": start_doc["timestamp"]}
        },
        sort=[("timestamp", 1)],
        projection={"timestamp": 1}
    )

    end_ts = datetime.fromisoformat(end_doc["timestamp"].replace("Z", "+00:00"))

    start_iso = start_ts.isoformat().replace("+00:00", "Z")
    end_iso   = end_ts.isoformat().replace("+00:00", "Z")

    count_targets = coll.count_documents({
        "game": "mhs",
        "playerId": pid,
        "eventType": "DialogueEvent",
        "eventKey": {"$in": TARGET_KEYS},
        "timestamp": {
            "$gte": start_iso,
            "$lte": end_iso
        }
    })

    if count_targets <= 1:
        green.append(pid)
    else:
        yellow.append(pid)

yellow = sorted(set(yellow))
green = sorted(set(green))
```
